# Data Visualization

```python
#Visualizing sales by product category using a bar plot
sns.barplot(x='Category', y='Sales', data=data)
plt.title('Sales by Product Category')
plt.show()
```

```python
#Visualizing monthly sales over time using a line plot

#data['Month'] = data['Order Date'].dt.to_period('M')

# Grouping by 'Month' and summing the 'Sales' column
monthly_sales = data.groupby('Month')['Sales'].sum()

# Converting index to timestamp
monthly_sales.index = monthly_sales.index.to_timestamp()

# Plotting the line graph
monthly_sales.plot(kind='line')
plt.title('Monthly Sales Over Time')
plt.show()
```

```python
#Visualizing sales vs profit using a scatter plot

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales', y='Profit', data=data, color='b', alpha=0.6)

# Adding titles and labels
plt.title('Sales vs Profit', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Profit', fontsize=12)

# Showing the plot
plt.grid(True)
plt.show()
```

```python
#Visualizing Sales by Region and Product Category using a heatmap

# Creating a pivot table to organize data by Region and Category, and aggregate sales values
sales_pivot = pd.pivot_table(data, values='Sales', index='Region', columns='Category', aggfunc='sum')

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, annot=True, fmt='.2f', cmap='Blues')

# Adding titles and labels
plt.title('Sales by Region and Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Region', fontsize=12)

# Showing the plot
plt.show()
```

```python
#Visuallizing Seasonal Sales trend using Time series Analysis

from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(monthly_sales, model='additive')

# Plotting the decomposed components
plt.figure(figsize=(12, 8))

# Plotting observed sales data
plt.subplot(411)
plt.plot(monthly_sales, label='Observed Sales')
plt.title('Observed Sales')

# Plotting trend component
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color='orange')
plt.title('Trend')

# Plotting seasonal component
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality', color='green')
plt.title('Seasonality')

# Plotting residual component (noise)
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual', color='red')
plt.title('Residuals')

plt.tight_layout()
plt.show()
```
